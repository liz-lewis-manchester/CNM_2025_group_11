```python
# import packages
import numpy as np
import matplotlib.pyplot as plt

# ask user to input spatial domain data
x_start = float(input("Enter start of distance domain (m): "))
x_end = float(input("Enter end of distance domain (m): "))
dx = float(input("Enter spatial resolution (m): "))

# ask user to input temporal domain data
t_start = float(input("Enter start of time domain (s): "))
t_end = float(input("Enter end of time domain (s): "))
dt = float(input("Enter temporal resolution dt (s): "))

# ask user to input flow speed data
U = float(input("Enter the flow speed (m/s): "))

# import concentration data from the initial conditions CSV file
file_name = input("Enter CSV filename for initial conditions: ")
file_data = np.loadtxt(file_name, delimiter=',', skiprows=1)
theta = file_data[:, 1]

# initialise distance domain
def initialise_distance(x_start, x_end, dx):
    x = np.arange(x_start, x_end + dx, dx)
    return x
x = initialise_distance(x_start, x_end, dx)

# initialise time domain
def initialise_time(t_start, t_end, dt):
    t = np.arange(t_start, t_end + dt, dt)
    return t
t = initialise_time(t_start, t_end, dt)

# finite difference method
def FDM(theta, U, dt, dx):
  new_theta = theta.copy()
  for i in range(1, len(theta)):
    new_theta[i] = theta[i] - U * (dt/dx) * (theta[i] - theta [i-1])
  new_theta[0] = theta[0]
  return new_theta

# simulation
results = [theta]

for n in range(1, len(t)):
    theta = FDM(theta, U, dt, dx)
    results.append(theta)
```
